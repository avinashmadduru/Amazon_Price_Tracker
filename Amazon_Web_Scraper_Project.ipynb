```python
import csv
import datetime
import os
import requests
from bs4 import BeautifulSoup


def check_price():
    URL = "https://www.amazon.in/Alienware-AW2525HM-FreeSync-Gameplay-DisplayPort/dp/B0FH5NMXHK/ref=sr_1_4?sr=8-4"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36"
    }

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")

    title_element = soup.find(id="productTitle")
    if not title_element:
        print("Could not find product title.")
        return
    title = title_element.get_text().strip()

    price_element = soup.find(class_="a-price-whole")
    if not price_element:
        print("Could not find price.")
        return
    price_raw = price_element.get_text()
    price = "".join(char for char in price_raw if char.isdigit())

    today = datetime.date.today()

    file_name = "AmazonWebScraperDataset.csv"
    file_exists = os.path.isfile(file_name)

    with open(file_name, "a", newline="", encoding="UTF-8") as f:
        writer = csv.writer(f)

        if not file_exists:
            writer.writerow(["Title", "Price", "Date"])

        writer.writerow([title, price, today])

    # Optional Email Alert
    # if price and int(price) <= 20000:
    #     send_mail()
```
